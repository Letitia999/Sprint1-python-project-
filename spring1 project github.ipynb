


#1.D     visual shape of the distribution
--final_standing

Mode- There appears to be a single mode ,The mode is located around the value  
      of 4.
Narrowness- Distribution is relatively narrow
Outliers- There are a few outliers.

  --   Home_Runs  
Symmetric- Distribution is roughly symmetrical,  there are about as many teams
          with below-average home run numbers as there are teams with above-average home run numbers.
Mode-There appears to be a single mode, which is the bin around 100 home runs.
Narrow - Distribution is relatively narrow,most teams have a number of home  
         runs that is close to the average.
Outliers- There are not many outliers

     --Runs_Scored
Symmetric-  symmetrical almost .
Mode- There appears to be a single mode around the 200-run .
Narrowness- The distribution is relatively narrow, There are not many scores
            that deviate significantly from the typical range.
Outliers- There are a few outliers.There are a couple of scores above 800 runs,
           which is much higher than the most common scores.

--Saves
Symmetric-  The distribution is left-skewed, meaning there are more saves with
            lower values than there are saves with higher values. This is because most teams tend to have a small number of saves, while a few teams have a very large number of saves

Mode- There is a single mode, which is around 10 saves.

Narrowness- The distribution is fairly wide, meaning that there is a lot of
             variability in the number of saves from team to team.

Outliers- There are a few outliers


--Complete_Games
Symmetric-  The distribution is left-skewed, meaning there are more complete
            games with fewer innings than there are complete games with more innings.
Mode- single mode at 7 innings
Narrowness- relatively narrow
Outliers- There are a few outliers, meaning there are a few games that were
            played for many more innings than usual.

--Strikeouts_Allowed

Symmetric- symmetrical, left and right sides mirror each other fairly closely
Mode- single mode, around the 600-strikeout mark
Narrowness- wider than a normal distribution
Outliers- several outliers

creating the plot

```python
corr_games_lost_runs_scored = df3.corr().loc['Games_Lost', 'Runs_Scored']
corr_games_lost_runs_against = df3.corr().loc['Games_Lost','Runs_Against']
df3['Runs_Scored_Minus_Runs_Against'] = df3['Runs_Scored'] - df3['Runs_Against']
corr_games_lost_runs_diff = df3.loc[:, 'Games_Lost'].corr(df3.loc[:, 'Runs_Scored_Minus_Runs_Against'])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(13, 3))

# Scatter plot Games Lost vs Runs Scored


ax[0].scatter(df3['Runs_Scored'], df3['Games_Lost'])

ax[0].set_xlabel('Runs_Scored', fontsize=10)
ax[0].set_ylabel('Games_Lost', fontsize=10)
ax[0].set_title('r=' + str(round(corr_games_lost_runs_scored,3)))

plt.show



# Scatter plot Games Lost vs Runs Against

ax[1].scatter(df3['Runs_Against'], df3['Games_Lost'])


ax[1].set_xlabel('Runs_Against', fontsize=10)
ax[1].set_ylabel('Games_Lost', fontsize=10)
ax[1].set_title('r=' + str(round(corr_games_lost_runs_against,3)))

plt.show



# Scatter plot Games Lost vs (Runs Scored minus Runs Against)

ax[2].scatter(df3['Runs_Scored_Minus_Runs_Against'], df3['Games_Lost'])


ax[2].set_xlabel('Runs_Scored_Minus_Runs_Against', fontsize=10)
ax[2].set_ylabel('Games_Lost', fontsize=10)
ax[2].set_title('r=' + str(round(corr_games_lost_runs_diff,3)))


plt.show
```
